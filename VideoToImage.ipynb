```python
import cv2
import random

video_path = '/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_artificial.MOV'
output_path = '/content/drive/MyDrive/MLandAIAssignment2/Images/Car_artificial/'
frame_count = 0
num_frames_to_extract = 10

# Open the video file
vidcap = cv2.VideoCapture(video_path)
success, image = vidcap.read()

# Get the total number of frames in the video
total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

# Create a list of random frame indices to extract
random_frame_indices = random.sample(range(total_frames), num_frames_to_extract)

while success and frame_count < total_frames:
    if frame_count in random_frame_indices:
        image_name = f'frame{frame_count}.PNG'
        cv2.imwrite(output_path + image_name, image)
        print('Saved frame:', frame_count)

    success, image = vidcap.read()
    frame_count += 1

# Release the video capture object
vidcap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os
import random

# Define a list of video file paths and corresponding output folder paths
video_paths = ['/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_artificial.MOV', '/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_shadow.MOV']
output_folders = ['/content/drive/MyDrive/MLandAIAssignment2/Images/Car_artificial/', '/content/drive/MyDrive/MLandAIAssignment2/Images/Car_shadow/']

num_frames_to_extract = 10

for video_path, output_path in zip(video_paths, output_folders):
    # Create the output folder if it doesn't exist
    os.makedirs(output_path, exist_ok=True)

    # Open the video file
    vidcap = cv2.VideoCapture(video_path)
    success, image = vidcap.read()

    # Get the total number of frames in the video
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Create a list of random frame indices to extract
    random_frame_indices = random.sample(range(total_frames), num_frames_to_extract)

    frame_count = 0

    while success:
        if frame_count in random_frame_indices:
            image_name = f'frame{frame_count}.PNG'
            image_path = os.path.join(output_path, image_name)
            cv2.imwrite(image_path, image)
            print(f'Saved frame {frame_count} from {video_path} to {output_path}')

        success, image = vidcap.read()
        frame_count += 1

    # Release the video capture object for this video
    vidcap.release()

# Optionally, close any open windows
cv2.destroyAllWindows()
```
